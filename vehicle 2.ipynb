```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import hvplot.pandas

from sklearn.model_selection import train_test_split

from sklearn import metrics

from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
df=pd.read_csv('./vehicle.csv')
```

```python
x=np.array(df[['year']])
Y=df['selling_price'].values
```

```python
modelo = LinearRegression()
modelo.fit(x, Y)
```

```python
print (f" la ecuacion de la recta es: y = {modelo.intercept_} + {modelo.coef_[0 ]}x")
```

```python
plt.scatter (x, Y, color='blue' , label='Datos')
plt.plot(x, modelo.predict(x), color='red' , label= 'linea de regresion')
plt.rcParams['figure.figsize']= (20, 9)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred =modelo.predict(x)
mse = mean_squared_error(Y, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(Y, y_pred)
r2= r2_score(Y, y_pred)
```

```python
print()
print(f" error cuadratico medio (MSE) : {mse}")
print()
print(f" raiz del error cuadratico medio (RMSE) :{rmse}")
print()
print(f" error absoluto medio (MAE) : {mae}")
print()
print(f" el coeficiente de determinacion (R*R): {r2}")
```
